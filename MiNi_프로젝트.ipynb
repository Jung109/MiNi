```python
### 학습용 데이터의 모양과 인덱스 확인

# 모양 shape
print(f"X_train의 모양:{X_train.shape}")
print('*'*88)

print(f'y_train의 모양:{y_train.shape}')
print('*'*88)

# 인덱스
print(f'X_train의 index:\n{X_train.index}')
print('*'*88)

print(f'y_train의 index:\n{y_train.index}')
print('*'*88)
```

```python
### 데이터 스케일링

# 필요한 함수 임폴트
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler 함수 호출
scaler=MinMaxScaler()

# 학습용 데이터 스케일링 실행
X_train_scaled=scaler.fit_transform(X_train)

# 평가용 데이터 스케일링 실행
X_test_scaled=scaler.transform(X_test)
print('*'*88)

# 결과 확인
print(f"학습용 데이터 스케일링 결과 확인:\n{X_train_scaled} ")
print('*'*88)

print(f'평가용 데이터 스케일링 결과 확인:\n{X_test_scaled}')
print('*'*88)
```
